九九乘法表

```python
for i in range(2,10):
    for j in range(1,10):
      product = i * j
      print('%d*%d=%2d' %(i,j,product),end=' ')
    print()
```

最小公倍數

```python
a = int(input('請輸入a的值: '))
b = int(input('請輸入b的值: '))

maxno = a * b
for i in range(1,maxno+1):
  if(i % a == 0 and i % b == 0):
    break
print('%d和%d的最小公倍數 = %d' %(a,b,i))
```

數列排除5

```python
n = int(input('請輸入一個正整數: '))
for i in range(1,n+1):
  if i % 5 == 0:
    continue
  print(i,end=' ')
```

綜合練習

```python
total = a = 0
while (a <= 50):
  total += a
  a += 2
print('0~50的偶數和結果是:',total,sep='')
加總 = 0
b = 1
while (b <= 50):
  加總 += b
  b += 2
print('0~50的奇數和結果是:',加總,sep='')
```

```python
a = 0
for b in range(1,101):
  if (b % 3 == 0 or b % 7 == 0):
    a += b
    b += 1
print('數字1~100中，所有是 3 或 7 倍數的數之總和=',a)
```

```python
a = int(input('請輸入正整數:'))
print('%d的因數有'%a,end=' ')
b = 1
c = 0
while b <= a:
  if (a % b == 0):
    c += 1
    if (a % b == 0):
      print(b,end=' ')
  b += 1
if c == 2:
  print('\n',a,'是質數',sep='')
else:
  print('\n',a,'不是質數',sep='')
```
